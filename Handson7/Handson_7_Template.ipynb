## 8.1 Tip Calculator
Add exception handling to a Tip Calculator program.

### Console:
```powershell
Tip Calculator

INPUT
Cost of meal: ten
Must be valid decimal number. Please try again. 
Cost of meal: -10
Must be greater than 0. Please try again. 
Cost of meal: 52.31
Tip percent:	17.5
Must be valid integer. Please try again. 
Tip percent:	20

OUTPUT
Cost of meal:   52.31 
Tip percent:    20%
Tip amount:     10.46
Total amount:   62.77
```

### Specifications:
- The program should accept **decimal** entries like 52.31 and 15.5 for the cost of the meal.
- The program should accept **integer** entries like 15, 20, 25 for the tip percent.
- The program should validate both user entries. That way, the user can’t crash the program by entering invalid data.
- The program should only accept numbers that are **greater than 0**.
- The program should round results to a maximum of two decimal places.


````python
def main():
    
    print("Tip Calculator")
    
    
    while True:
        try:
            cost = float(input("\nCost of Meal: "))
            if cost < 0:
                print("Must be greather than 0. Please try again.")
                continue
        except ValueError:
            print("Must be a valid decimal number. Please try again.")
        else:
            break
    
    while True:
        try:
            tipPercent = int(input("Tip Percent: "))
        except ValueError:
            print("Must be a valid integer. Please try again.")
        else:
            break
    
    
    
    print("\nCost of Meal:", cost)
    print("Tip Percent: " + str(tipPercent) + "%")
    tip = cost * (tipPercent/100)
    total = cost + tip
    print("Tip amount: ", round(tip, 2))
    print("Total amount: ", round(total,2))
    
    
if __name__ == "__main__":
    main()
````

## 8.2 Wizard Inventory

Add exception handling to a program that keeps track of the inventory of items that a wizard can carry. If you still have Wizard Inventory from Handson_5, you can add the exception handling to that program. Otherwise, you can start this program from scratch.

### Console if the program can't find the **inventory** file:


### The error message if the program can’t find the **items** file:
```powershell

Exiting program. Bye!	
```

### Specifications:
- This program should read the text file named `wizard_all_items.txt` that contains all the items a wizard can carry.
- When the user selects the walk command, the program should randomly pick one of the items that were read from the text file and give the user the option to grab it.
- The current items that the wizard is carrying should be saved in an inventory file. Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry four items at a time. For the drop command, display an error message if the user enters an invalid integer or an integer that doesn’t correspond with an item.
- Handle all exceptions that might occur so that the user can’t cause the program to crash. If the all items file is missing, display an appropriate error message and exit the program.
- If the inventory file is missing, display an appropriate error message and continue with an empty inventory for the user. That way, the program will write a new inventory file when the user adds items to the inventory.

````python
import random
import sys



def main():
    inventory = []
    items = []
    displayMenu()
    
    setItemOpts(items)
    setInventory(inventory)
    
    command = input("\nCommand: ")
    while command != 'exit':
        if command == 'walk':
            item = walk()
            grabItem(item, inventory)
        elif command == 'show':
            showItems(inventory)
        elif command == 'drop':
            dropItem(inventory)
        else:
            print("Invalid command. Try again")
        
        command = input("\nCommand: ")
        
    print("Bye!")
    
    
def setInventory(inventory):
    try:
        with open('wizard_inventory.txt', 'r', newline = '') as file:
            items = file.read().splitlines()
        inventory += items
    except FileNotFoundError:
        print("\nCould not find inventory file!")
        print("Wizard is starting with no inventory!")
        
def setItemOpts(items):
    try:
        with open('wizard_all_items.txt', 'r', newline = '') as file:
            items = file.read().splitlines()
        items += items
    except FileNotFoundError:
        print("\nCould not find items file.")
        print("Exiting file. Bye!")
        sys.exit()
  
      
def displayMenu():
    '''displays the title and command options for the program'''
    print("The Wizard Inventory Menu")
    print("\nCOMMAND MENU", "\nwalk - Walk down the path", "\nshow - Show all items", 
          "\ndrop - Drop an item", "\nexit - Exit program")


def walk():
    '''returns a random item from the list of possible items in wizard_all_items file'''
    with open('wizard_all_items.txt', 'r', newline = '') as file:
        items = file.read().splitlines()
        random.seed()
        x = random.randint(0, len(items)-1)
        item = items[x]
        return item
        

def grabItem(item, inventory):
    '''takes parameter of an item
    presents user with the item and gives the user the option to take the item
    if user takes item and has room, updates the file containing the inventory'''
    
    take = input('While walking down a path, you see {}. Do you want to grab it? (y/n): '.format(item))
    
    if take.lower() == 'y':
        if len(inventory) == 4:
            print("You can't carry anymore items. Drop something first.")
        else:
            inventory.append(item)
            print("You picked up", item)
            with open('wizard_inventory.txt', 'w', newline = '') as file:
                for i in inventory:
                    file.write(i + '\n')


def dropItem(inventory):
    '''allows the user to remove an item from their inventory
    updates inventory file when item is removed'''
    
    while True:
        try:
            remove = int(input("Number: "))
            while remove > len(inventory):
                remove = int(input("Invalid inventory position. Try again: "))
        except ValueError:
            print("Invalid Item Number.")
        else:
            break
    print("You dropped", inventory[remove-1])
    inventory.pop(remove-1)
        
    with open('wizard_inventory.txt', 'w', newline = '') as file:
        for item in inventory:
            file.write(item + '\n')


def showItems(inventory):
    '''shows the list of items the wizard currently has'''
    count = 1

    for item in inventory:
        print(str(count) + '. ' + item)
        count += 1
    

if __name__ == "__main__":
    main()
````
