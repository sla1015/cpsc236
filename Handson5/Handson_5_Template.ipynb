## 7.2 Wizard Inventory
Create a program that keeps track of the items that a wizard can carry.

### Console:
```powershell
The Wizard Inventroy Program

COMMAND MENU
walk - Walk down the path show - Show all items drop - Drop an item
exit - Exit program

Command: walk
While walking down a path, you see a scroll of uncursing. Do you want to grab it? (y/n): y
You picked up a scroll of uncursing.

Command: walk
While walking down a path, you see an unknown potion. Do you want to grab it? (y/n): y
You can't carry any more items. Drop something first.

Command: show
1.	a wooden staff
2.	a scroll of invisibility
3.	a crossbow
4.	a scroll of uncursing

Command: drop Number: 3
You dropped a crossbow.

Command: exit Bye!
```

### Specifications:
- Use the attached text file named `wizard_all_items.txt` that contains a list of all the items that a wizard can carry. (in the corresponding directory)
- When the user selects the walk command, the program should read the items from the file, randomly pick one, and give the user the option to grab it.
- Your program should create another file that stores the items that the wizard is carrying. Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry four items at a time.
- For the drop command, display an error message if the user enters an invalid number for the item.



````python
import random


def main():
    displayMenu()
    command = input("\nCommand: ")
    while command != 'exit':
        if command == 'walk':
            item = walk()
            grabItem(item)
        elif command == 'show':
            showItems()
        elif command == 'drop':
            dropItem()
        else:
            print("Invalid command. Try again")
        
        command = input("\nCommand: ")
        
    print("Bye!")
  
      
def displayMenu():
    '''displays the title and command options for the program'''
    print("The Wizard Inventory Menu")
    print("\nCOMMAND MENU", "\nwalk - Walk down the path", "\nshow - Show all items", 
          "\ndrop - Drop an item", "\nexit - Exit program")


def walk():
    '''returns a random item from the list of possible items in wizard_all_items file'''
    with open('wizard_all_items.txt', 'r', newline = '') as file:
        items = file.read().splitlines()
        random.seed()
        x = random.randint(0, len(items)-1)
        item = items[x]
        return item
        

def grabItem(item):
    '''takes parameter of an item
    presents user with the item and gives the user the option to take the item
    if user takes item and has room, updates the file containing the inventory'''
    take = input('While walking down a path, you see {}. Do you want to grab it? (y/n): '.format(item))
    with open('wizard_inventory.txt', 'r', newline = '') as file:
        items = file.read().splitlines()
    
    if take.lower() == 'y':
        if len(items) == 4:
            print("You can't carry anymore items. Drop something first.")
        else:
            items.append(item)
            print("You picked up", item)
            with open('wizard_inventory.txt', 'w', newline = '') as file:
                for i in items:
                    file.write(i + '\n')


def dropItem():
    '''allows the user to remove an item from their inventory
    updates inventory file when item is removed'''
    with open('wizard_inventory.txt', 'r', newline = '') as file:
        items = file.read().splitlines()
        remove = int(input("Number: "))
        while remove > len(items):
            remove = int(input("Invalid inventory position. Try again: "))
        print("You dropped", items[remove-1])
        items.pop(remove-1)
    with open('wizard_inventory.txt', 'w', newline = '') as file:
        for item in items:
            file.write(item + '\n')


def showItems():
    '''shows the list of items the wizard currently has'''
    count = 1
    with open('wizard_inventory.txt', 'r', newline = '') as file:
        items = file.read().splitlines()
        for item in items:
            print(str(count) + '. ' + item)
            count += 1
    


if __name__ == "__main__":
    main()
````

## 7.3 Monthly Sales
Create a program that reads the sales for 12 months from a file and calculates the total yearly sales as well as the average monthly sales. In addition, this program should let the user edit the sales for any month.

### Console:

### Specifications:
- Use the attached text file named CSV file named `monthly_sales.csv` that contains the month and sales data shown above.
- For the edit command, display an error message if the user doesn’t enter a valid three-letter abbreviation for the month.
- When the user edits the sales amount for a month, the data should be saved to the CSV file immediately. That way, no data is lost, even if the program crashes later.
- Round the results of the monthly average to a maximum of 2 decimal digits.

````python
import csv


def main():
    salesList = []
    salesMenu()
    with open('monthly_sales.csv', newline = '') as file:
        reader = csv.reader(file)
        for row in reader:
            salesList.append(row)
    
    while True:
        command = input("\nCommand: ")
        if command == 'monthly':
            viewMonths(salesList)
        elif command == 'yearly':
            viewYear(salesList)
        elif command == 'edit':
            editMonth(salesList)
        elif command == 'exit':
            break
        else:
            print("Invalid command. Try again.")
    
    print("Bye!")
            



def salesMenu():
    '''displays program title and command options'''
    print('Monthly Sales Program')
    print('\nCOMMAND MENU', '\nmonthly - View monthly sales',
          '\nyearly - View yearly summary', '\nedit - Edit sales for a month',
          '\nexit - Exit program')
    
def viewMonths(salesList):
    '''displays the sales for each month'''
    for month in salesList:
        print(month[0] + ' - ' + month[1])


def viewYear(salesList):
    '''displays total yearly sales and average sales per month'''
    totalSales = 0
    for month in salesList:
        totalSales += int(month[1])
    avgSales = totalSales/12
    avgSales = round(avgSales, 2)
    print("Yearly total: ", totalSales)
    print("Monthly average: ", avgSales)
    

def editMonth(salesList):
    '''allows individual monthly sales totals to be edited'''
    monthIn = input("Three-Letter Month: ")
    x = 13
    for row in salesList:
        if row[0] == monthIn:
            x = salesList.index(row)
    if x == 13:
        print("Invalid Abreviation. Month not found.")
        return
    month = salesList[x]
    newSales = input("Sales Amount: ")
    month[1] = newSales
    salesList[x] = month
    with open('monthly_sales.csv', 'w', newline = '') as file:
        writer = csv.writer(file)
        writer.writerows(salesList)
    print("Sales amount for {} was modified".format(month[0]))
    


if __name__ == "__main__":
    main()
````
